```python
import pandas as pd

# Load the dataframe
f42to57_train = pd.read_parquet('/content/drive/MyDrive/Transformed_Train_Data/f42to57_train.parquet')

# Calculate the null percentage for each column
null_percentage = f42to57_train.isnull().sum() / len(f42to57_train) * 100

# Display the null percentages
print("Null percentage for each column:")
print(null_percentage)
```

```python
import os
import pandas as pd

directory_path = '/content/drive/MyDrive/Transformed_Train_Data'
loaded_dataframes = {}

# List files in the directory
file_list = os.listdir(directory_path)

print(f"Loading files from: {directory_path}")

# Iterate through the files and load them
for filename in file_list:
    file_path = os.path.join(directory_path, filename)
    name, extension = os.path.splitext(filename)

    try:
        if extension == '.csv':
            loaded_dataframes[name] = pd.read_csv(file_path)
            print(f"Loaded {filename} as a CSV.")
        elif extension == '.parquet':
            loaded_dataframes[name] = pd.read_parquet(file_path)
            print(f"Loaded {filename} as a Parquet file.")
        else:
            print(f"Skipped {filename}: Unsupported file type.")
    except Exception as e:
        print(f"Error loading {filename}: {e}")

# List the names of the loaded dataframes
print("\nLoaded dataframes:")
for df_name in loaded_dataframes.keys():
    print(df_name)
```

```python
# Rename the dataframes in the dictionary
renamed_dataframes = {}
for name, df in loaded_dataframes.items():
    # Assuming the naming convention is consistent, we can extract the relevant part
    # This is a simplified approach and might need adjustment based on actual filenames
    new_name = name.replace('_processed', '').replace('_imputed', '').replace('_encoded', '')
    renamed_dataframes[new_name] = df

# Update the loaded_dataframes dictionary with the renamed dataframes
loaded_dataframes = renamed_dataframes

# List the names of the renamed dataframes
print("Renamed dataframes:")
for df_name in loaded_dataframes.keys():
    print(df_name)
```

```python
# Drop 'id2' column from f42to57_train
f42to57_train_dropped_id2 = f42to57_train.drop('id2', axis=1)

# Display shape after dropping column
print("Shape of f42to57_train after dropping 'id2':", f42to57_train_dropped_id2.shape)

# Merge f42to57_train_dropped_id2 with f1to41_train
merged_df = pd.merge(loaded_dataframes['f1to41_train'], f42to57_train_dropped_id2, left_index=True, right_index=True)

# Display shape after merging
print("Shape of merged dataframe:", merged_df.shape)
```

```python
# Add the merged dataframe to the loaded_dataframes dictionary with the new name
loaded_dataframes['f1to57_train'] = merged_df

# Display the names of the updated loaded dataframes
print("Updated loaded dataframes:")
for df_name in loaded_dataframes.keys():
    print(df_name)
```

```python
# Get the dataframes from the dictionary
f58to309_train_df = loaded_dataframes['f58to309_train']
f1to57_train_df = loaded_dataframes['f1to57_train']

# Identify the ID column to drop
id_column_to_drop = None
if 'id2' in f58to309_train_df.columns:
    id_column_to_drop = 'id2'
elif 'id3' in f58to309_train_df.columns:
    id_column_to_drop = 'id3'
else:
    print("Neither 'id2' nor 'id3' found in f58to309_train.")

if id_column_to_drop:
    # Display shapes before dropping
    print("Shape of f58to309_train before dropping 'id' column:", f58to309_train_df.shape)
    print("Shape of f1to57_train before merging:", f1to57_train_df.shape)

    # Drop the identified ID column
    f58to309_train_dropped_id = f58to309_train_df.drop(id_column_to_drop, axis=1)

    # Display shape after dropping
    print(f"Shape of f58to309_train after dropping '{id_column_to_drop}':", f58to309_train_dropped_id.shape)

    # Merge the dataframes
    f1to309_train = pd.merge(f1to57_train_df, f58to309_train_dropped_id, left_index=True, right_index=True)

    # Display shape after merging
    print("Shape of merged dataframe (f1to309_train):", f1to309_train.shape)

    # Add the merged dataframe to the loaded_dataframes dictionary with the new name
    loaded_dataframes['f1to309_train'] = f1to309_train

    # Display the names of the updated loaded dataframes
    print("\nUpdated loaded dataframes:")
    for df_name in loaded_dataframes.keys():
        print(df_name)
else:
    print("Merging could not be performed as no ID column was found to drop.")
```

```python
# Calculate the null percentage for each column in f1to309_train
f1to309_null_percentage = f1to309_train.isnull().sum() / len(f1to309_train) * 100

# Check for columns with null percentage greater than 0
columns_with_nulls = f1to309_null_percentage[f1to309_null_percentage > 0]

# Display the result
if not columns_with_nulls.empty:
    print("Columns in f1to309_train with null percentage greater than 0:")
    print(columns_with_nulls)
else:
    print("No columns in f1to309_train have a null percentage greater than 0.")
```

```python
# Dataframes to process in ascending order
dfs_to_merge_names = [
    'f310tof314_train',
    'f315tof319_train',
    'f320tof324_train',
    'f325to328_train',
    'f329to331_train',
    'f332to334_train'
]

# Start with the f1to309_train dataframe as the base for merging
merged_f1to309_to_f334 = loaded_dataframes['f1to309_train']
print("Initial shape of base dataframe (f1to309_train):", merged_f1to309_to_f334.shape)

for df_name in dfs_to_merge_names:
    print(f"\nProcessing {df_name}...")
    current_df = loaded_dataframes[df_name]

    # Identify the ID column to drop
    id_column_to_drop = None
    if 'id2' in current_df.columns:
        id_column_to_drop = 'id2'
    elif 'id3' in current_df.columns:
        id_column_to_drop = 'id3'
    else:
        print(f"Neither 'id2' nor 'id3' found in {df_name}. Skipping column drop for this dataframe.")

    if id_column_to_drop:
        # Display shape before dropping
        print(f"Shape of {df_name} before dropping '{id_column_to_drop}':", current_df.shape)
        # Drop the identified ID column
        current_df_dropped_id = current_df.drop(id_column_to_drop, axis=1)
        # Display shape after dropping
        print(f"Shape of {df_name} after dropping '{id_column_to_drop}':", current_df_dropped_id.shape)
    else:
        current_df_dropped_id = current_df # No column to drop, use the original dataframe

    # Display shape of base dataframe before merging
    print(f"Shape of base dataframe before merging with {df_name}:", merged_f1to309_to_f334.shape)
    # Merge with the base dataframe
    merged_f1to309_to_f334 = pd.merge(merged_f1to309_to_f334, current_df_dropped_id, left_index=True, right_index=True)

    # Display shape after merging
    print(f"Shape of base dataframe after merging with {df_name}:", merged_f1to309_to_f334.shape)

# The final merged dataframe is stored in merged_f1to309_to_f334
loaded_dataframes['f1to334_train'] = merged_f1to309_to_f334

print("\nFinal merged dataframe name: f1to334_train")
print("Final shape of f1to334_train:", loaded_dataframes['f1to334_train'].shape)

# Display the names of the updated loaded dataframes
print("\nUpdated loaded dataframes:")
for df_name in loaded_dataframes.keys():
    print(df_name)
```

```python
# Drop the specified columns from f1to366_train
columns_to_drop_f1to366 = ['id4', 'f332', 'f333', 'f334']
f1to366_train_dropped = loaded_dataframes['f1to366_train'].drop(columns=columns_to_drop_f1to366)

# Display the shape of the dataframe after dropping columns
print("Shape of f1to366_train after dropping columns:", f1to366_train_dropped.shape)

# Display the first few rows of the modified dataframe
print("\nFirst 5 rows of the modified f1to366_train dataframe:")
display(f1to366_train_dropped.head())
```

```python
# Create a copy to avoid modifying the original dataframe in place
f332to340_train = merged_f332_to_f340.copy()

# Identify columns for imputation
columns_to_impute = ['f332', 'f333', 'f334']

# Calculate global medians for fallback
global_medians = f332to340_train[columns_to_impute].median()

# Impute null values for each column
for col in columns_to_impute:
    # Fill with group median first
    f332to340_train[col] = f332to340_train.groupby('id2')[col].transform(lambda x: x.fillna(x.median()))

    # Fill remaining nulls with global median
    f332to340_train[col].fillna(global_medians[col], inplace=True)

# Display the null percentage after imputation
print("Null percentage for each column in f332to340_train after imputation:")
print(f332to340_train.isnull().sum() / len(f332to340_train) * 100)

# Display the first few rows of the imputed dataframe
print("\nFirst 5 rows of the imputed dataframe:")
display(f332to340_train.head())
```

```python
# Display shape of f332to340_train before dropping
print("Shape of f332to340_train before dropping 'id2':", f332to340_train.shape)

# Drop 'id2' column from f332to340_train
f332to340_train_dropped_id2 = f332to340_train.drop('id2', axis=1)

# Display shape after dropping
print("Shape of f332to340_train after dropping 'id2':", f332to340_train_dropped_id2.shape)

# Display shape of f1to366_train before merging
print("Shape of f1to366_train before merging:", loaded_dataframes['f1to366_train'].shape)

# Merge f1to366_train with f332to340_train_dropped_id2
merged_f1to366_f332to340 = pd.merge(loaded_dataframes['f1to366_train'], f332to340_train_dropped_id2, left_index=True, right_index=True)

# Display shape after merging
print("Shape of merged dataframe:", merged_f1to366_f332to340.shape)
```

```python
# Display shape of f332to340_train before dropping
print("Shape of f332to340_train before dropping 'id2':", f332to340_train.shape)

# Drop 'id2' column from f332to340_train
f332to340_train_dropped_id2 = f332to340_train.drop('id2', axis=1)

# Display shape after dropping
print("Shape of f332to340_train after dropping 'id2':", f332to340_train_dropped_id2.shape)

# Display shape of f1to366_train_dropped before merging
print("Shape of f1to366_train_dropped before merging:", f1to366_train_dropped.shape)

# Merge f1to366_train_dropped with f332to340_train_dropped_id2
final_train = pd.merge(f1to366_train_dropped, f332to340_train_dropped_id2, left_index=True, right_index=True)

# Display shape after merging
print("Shape of final_train after merging:", final_train.shape)
```

```python
import pandas as pd
import os

test_files_to_load = [
    '/content/drive/MyDrive/Transformed_Test_Data/f310tof314_test_processed.parquet',
    '/content/drive/MyDrive/Transformed_Test_Data/f315tof319_test_processed.parquet',
    '/content/drive/MyDrive/Transformed_Test_Data/f320tof324_test_processed.parquet',
    '/content/drive/MyDrive/Transformed_Test_Data/f325to328_test_processed.parquet',
    '/content/drive/MyDrive/Transformed_Test_Data/f329to331_test_processed.parquet'
]

loaded_test_dataframes = {}

print("Loading test dataframes:")

for file_path in test_files_to_load:
    try:
        # Extract filename and remove extension and '_processed'
        filename = os.path.basename(file_path)
        name, extension = os.path.splitext(filename)
        new_name = name.replace('_processed', '')

        loaded_test_dataframes[new_name] = pd.read_parquet(file_path)
        print(f"Loaded {filename} as {new_name}")

    except FileNotFoundError:
        print(f"Error: File not found at {file_path}")
    except Exception as e:
        print(f"Error loading {file_path}: {e}")

print("\nLoaded test dataframes:")
for df_name in loaded_test_dataframes.keys():
    print(df_name)
```

```python
# Dataframes to process in ascending order
test_dfs_to_merge_names = [
    'f310tof314_test',
    'f315tof319_test',
    'f320tof324_test',
    'f325to328_test',
    'f329to331_test'
]

# Start with the f1to309_test dataframe as the base for merging
merged_f1to309_to_f331_test = f1to309_test.copy() # Create a copy to avoid modifying the original f1to309_test
print("Initial shape of base dataframe (f1to309_test):", merged_f1to309_to_f331_test.shape)

for df_name in test_dfs_to_merge_names:
    print(f"\nProcessing {df_name}...")
    current_df = loaded_test_dataframes[df_name]

    # Identify the ID column to drop
    id_column_to_drop = None
    if 'id2' in current_df.columns:
        id_column_to_drop = 'id2'
    elif 'id3' in current_df.columns:
        id_column_to_drop = 'id3'
    else:
        print(f"Neither 'id2' nor 'id3' found in {df_name}. Skipping column drop for this dataframe.")

    if id_column_to_drop:
        # Display shape before dropping
        print(f"Shape of {df_name} before dropping '{id_column_to_drop}':", current_df.shape)
        # Drop the identified ID column
        current_df_dropped_id = current_df.drop(id_column_to_drop, axis=1)
        # Display shape after dropping
        print(f"Shape of {df_name} after dropping '{id_column_to_drop}':", current_df_dropped_id.shape)
    else:
        current_df_dropped_id = current_df # No column to drop, use the original dataframe

    # Display shape of base dataframe before merging
    print(f"Shape of base dataframe before merging with {df_name}:", merged_f1to309_to_f331_test.shape)
    # Merge with the base dataframe (assuming merging on index is appropriate after resetting indices earlier)
    merged_f1to309_to_f331_test = pd.merge(merged_f1to309_to_f331_test, current_df_dropped_id, left_index=True, right_index=True)

    # Display shape after merging
    print(f"Shape of base dataframe after merging with {df_name}:", merged_f1to309_to_f331_test.shape)

# The final merged dataframe is stored in merged_f1to309_to_f331_test
f1to331_test = merged_f1to309_to_f331_test

print("\nFinal merged dataframe name: f1to331_test")
print("Final shape of f1to331_test:", f1to331_test.shape)
```

```python
import pandas as pd

# Load the dataframe
test_data_f336to340 = pd.read_parquet('/content/drive/MyDrive/Transformed_Test_Data/test_data_f336to340_imputed.parquet')

# Identify columns to keep
columns_to_keep_test = ['id2'] + [f'f{i}' for i in range(336, 341)]

# Drop columns not in the list of columns to keep
columns_to_drop_test = [col for col in test_data_f336to340.columns if col not in columns_to_keep_test]
f336to340_test = test_data_f336to340.drop(columns=columns_to_drop_test)

# Display the shape of the processed dataframe
print("Shape of f336to340_test after dropping columns:", f336to340_test.shape)

# Display the first few rows of the processed dataframe
print("\nFirst 5 rows of f336to340_test:")
print(f336to340_test.head())

# Calculate the null percentage for each column in f336to340_test
f336to340_test_null_percentage = f336to340_test.isnull().sum() / len(f336to340_test) * 100

# Display the null percentages
print("\nNull percentage for each column in f336to340_test:")
print(f336to340_test_null_percentage)
```

```python
# Create a copy to avoid modifying the original dataframe in place if needed later
f332to334_test_imputed = f332to334_test.copy()

# Identify columns for imputation
columns_to_impute_test = ['f332', 'f333', 'f334']

# Calculate global medians for fallback
global_medians_test = f332to334_test_imputed[columns_to_impute_test].median()

# Impute null values for each column
for col in columns_to_impute_test:
    # Fill with group median first
    f332to334_test_imputed[col] = f332to334_test_imputed.groupby('id2')[col].transform(lambda x: x.fillna(x.median()))

    # Fill remaining nulls with global median
    f332to334_test_imputed[col].fillna(global_medians_test[col], inplace=True)

# Assign the imputed dataframe back to f332to334_test (as requested)
f332to334_test = f332to334_test_imputed

# Display the null percentage after imputation
print("Null percentage for each column in f332to334_test after imputation:")
print(f332to334_test.isnull().sum() / len(f332to334_test) * 100)

# Display the first few rows of the imputed dataframe
print("\nFirst 5 rows of the imputed f332to334_test dataframe:")
print(f332to334_test.head())
```

```python
# Drop 'id2' from f336to340_test
f336to340_test_dropped_id2 = f336to340_test.drop('id2', axis=1)
print("Shape of f336to340_test after dropping 'id2':", f336to340_test_dropped_id2.shape)

# Drop 'id2' from f332to334_test
f332to334_test_dropped_id2 = f332to334_test.drop('id2', axis=1)
print("Shape of f332to334_test after dropping 'id2':", f332to334_test_dropped_id2.shape)

# Reset index for merging without index
f1to366_test_reset = f1to366_test.reset_index(drop=True)
f336to340_test_reset = f336to340_test_dropped_id2.reset_index(drop=True)
f332to334_test_reset = f332to334_test_dropped_id2.reset_index(drop=True)

print(f"\nShape of f1to366_test after index reset: {f1to366_test_reset.shape}")
print(f"Shape of f336to340_test after index reset: {f336to340_test_reset.shape}")
print(f"Shape of f332to334_test after index reset: {f332to334_test_reset.shape}")


# Merge f1to366_test and f336to340_test
print("\nMerging f1to366_test and f336to340_test...")
print(f"Shape of f1to366_test before merging: {f1to366_test_reset.shape}")
print(f"Shape of f336to340_test before merging: {f336to340_test_reset.shape}")
merged_f1to366_f336to340 = pd.concat([f1to366_test_reset, f336to340_test_reset], axis=1)
print(f"Shape after merging f1to366_test and f336to340_test: {merged_f1to366_f336to340.shape}")

# Merge the result with f332to334_test
print("\nMerging with f332to334_test...")
print(f"Shape before merging with f332to334_test: {merged_f1to366_f336to340.shape}")
print(f"Shape of f332to334_test before merging: {f332to334_test_reset.shape}")
final_test = pd.concat([merged_f1to366_f336to340, f332to334_test_reset], axis=1)
print(f"Shape of final_test after merging: {final_test.shape}")

# Name the final dataframe
# The dataframe is already named final_test by assignment

print("\nFinal dataframe named: final_test")
```
