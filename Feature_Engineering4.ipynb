```python
# Define the columns that require imputation
imputation_cols = [f'f{i}' for i in range(198, 213)]
print("Columns to be imputed:", imputation_cols)
```

**Reasoning**:
Implement the corrected column-wise imputation for the training data (`f198to212_train`) using the provided strategy.
This involves iterating through each imputation column, calculating the group median for that column, filling nulls with the group median, and then using the global median as a fallback for any remaining nulls.

```python
# precompute global medians on non-null values for training data
global_medians_train = f198to212_train[imputation_cols].dropna().median()

for col in imputation_cols:
    # 1) fillna by customer‐group median
    filled = (
        f198to212_train
        .groupby('id2')[col]
        .transform(lambda s: s.fillna(s.median()))
    )
    # 2) fill any remaining with the global median for that col
    f198to212_train[col] = filled.fillna(global_medians_train[col])

# Check null percentages after imputation for training data
print("Null % after imputation for f198to212_train:")
print(f198to212_train[imputation_cols].isnull().mean() * 100)
```

**Reasoning**:

This involves iterating through each imputation column, calculating the group median for that column, filling nulls with the group median, and then using the global median as a fallback for any remaining nulls.

```python
# precompute global medians on non-null values for testing data
global_medians_test = f198to212_test[imputation_cols].dropna().median()

for col in imputation_cols:
    # 1) fillna by customer‐group median
    filled = (
        f198to212_test
        .groupby('id2')[col]
        .transform(lambda s: s.fillna(s.median()))
    )
    # 2) fill any remaining with the global median for that col
    f198to212_test[col] = filled.fillna(global_medians_test[col])

# Check null percentages after imputation for testing data
print("Null % after imputation for f198to212_test:")
print(f198to212_test[imputation_cols].isnull().mean() * 100)
```

```python
import pandas as pd
import os

# Define the directory path
data_dir = '/content/drive/MyDrive/Colab Notebooks/AMEXChallenge'

# Define the filenames for train and test data
train_file = os.path.join(data_dir, 'train_data.parquet')
test_file = os.path.join(data_dir, 'test_data.parquet')

# Load the train and test data
try:
    train_df = pd.read_parquet(train_file)
    test_df = pd.read_parquet(test_file)

    print("Train data loaded successfully:")
    display(train_df.head())

    print("\nTest data loaded successfully:")
    display(test_df.head())

except FileNotFoundError:
    print(f"Error: Make sure the files '{os.path.basename(train_file)}' and '{os.path.basename(test_file)}' exist in the directory '{data_dir}'")
except Exception as e:
    print(f"An error occurred while loading the data: {e}")
```

```python
# Select 'id2', 'id4', and columns from 'f142' to 'f174' for the test data
f142to174_test = test_df[['id2', 'id4'] + [f'f{i}' for i in range(142, 174)]].copy()

# Display the head of the new test dataframe
print("\nf142to174_test dataframe:")
display(f142to174_test.head())
```

```python
# Convert 'id4' to datetime objects
f142to174_train['id4'] = pd.to_datetime(f142to174_train['id4'])
f142to174_test['id4'] = pd.to_datetime(f142to174_test['id4'])

# Extract the hour and create a new 'hours' column
f142to174_train['hours'] = f142to174_train['id4'].dt.hour
f142to174_test['hours'] = f142to174_test['id4'].dt.hour

print("f142to174_train with 'hours' column:")
display(f142to174_train.head())

print("\nf142to174_test with 'hours' column:")
display(f142to174_test.head())
```

**Reasoning**:
The subtask requires dropping the 'id4' column from both dataframes and displaying the head of the modified dataframes.



```python
f142to174_train = f142to174_train.drop('id4', axis=1)
f142to174_test = f142to174_test.drop('id4', axis=1)

print("f142to174_train after dropping 'id4':")
display(f142to174_train.head())

print("\nf142to174_test after dropping 'id4':")
display(f142to174_test.head())
```

**Reasoning**:
Convert all columns except 'id2' to numeric in both dataframes and display the data types to verify.



```python
for col in f142to174_train.columns:
    if col != 'id2':
        f142to174_train[col] = pd.to_numeric(f142to174_train[col], errors='coerce')

for col in f142to174_test.columns:
    if col != 'id2':
        f142to174_test[col] = pd.to_numeric(f142to174_test[col], errors='coerce')

print("Data types for f142to174_train after numeric conversion:")
display(f142to174_train.dtypes)

print("\nData types for f142to174_test after numeric conversion:")
display(f142to174_test.dtypes)
```

## Calculate null percentage

### Subtask:
Calculate the percentage of null values for each column in both dataframes.


**Reasoning**:
Calculate and display the percentage of null values for each column in both dataframes.



```python
null_percentage_train = (f142to174_train.isnull().sum() / len(f142to174_train)) * 100
null_percentage_test = (f142to174_test.isnull().sum() / len(f142to174_test)) * 100

print("Null percentage for f142to174_train:")
display(null_percentage_train)

print("\nNull percentage for f142to174_test:")
display(null_percentage_test)
```

**Reasoning**:
Calculate the range (min and max) for the numeric columns in both dataframes.



```python
range_train = f142to174_train.describe().loc[['min', 'max']]
range_test = f142to174_test.describe().loc[['min', 'max']]

print("Range (Min and Max) for f142to174_train:")
display(range_train)

print("\nRange (Min and Max) for f142to174_test:")
display(range_test)
```

```python
import numpy as np

# Impute 'f168' in f142to174_train
# Calculate global median for fallback in training data
global_median_f168_train = f142to174_train['f168'].median()

# Fill NaNs with group median, then global median for training data
f142to174_train['f168'] = f142to174_train.groupby('hours')['f168'].transform(lambda x: x.fillna(x.median()))
f142to174_train['f168'] = f142to174_train['f168'].fillna(global_median_f168_train)

print("Null percentage in f168 after imputation for f142to174_train:")
print(f142to174_train['f168'].isnull().mean() * 100)

# Impute 'f168' in f142to174_test
# Calculate global median for fallback in testing data
global_median_f168_test = f142to174_test['f168'].median()

# Fill NaNs with group median, then global median for testing data
f142to174_test['f168'] = f142to174_test.groupby('hours')['f168'].transform(lambda x: x.fillna(x.median()))
f142to174_test['f168'] = f142to174_test['f168'].fillna(global_median_f168_test)

print("\nNull percentage in f168 after imputation for f142to174_test:")
print(f142to174_test['f168'].isnull().mean() * 100)
```

**Reasoning**:
Create the lists of column names for imputation based on the specified ranges and exclusion.



```python
# Create a list for columns f142 to f145
imputation_cols_142_145 = [f'f{i}' for i in range(142, 146)]

# Create a list for columns f146 to f151
imputation_cols_146_151 = [f'f{i}' for i in range(146, 152)]

# Create a list for columns f152 to f173, excluding f168
imputation_cols_152_173 = [f'f{i}' for i in range(152, 174) if f'f{i}' != 'f168']

# Print the lists
print("Imputation columns f142-f145:", imputation_cols_142_145)
print("Imputation columns f146-f151:", imputation_cols_146_151)
print("Imputation columns f152-f173 (excluding f168):", imputation_cols_152_173)
```

**Reasoning**:
The error indicates that columns 'f144' and 'f145' are not present in the dataframe, likely because they were dropped in a previous step. I need to remove these columns from the `imputation_cols_142_145` list before proceeding with imputation.



```python
# Remove f144 and f145 from the imputation_cols_142_145 list
imputation_cols_142_145 = [col for col in imputation_cols_142_145 if col not in ['f144', 'f145']]

# Combine all imputation columns into a single list for easier iteration
all_imputation_cols = imputation_cols_142_145 + imputation_cols_146_151 + imputation_cols_152_173

# Precompute global medians on non-null values for training data
global_medians_train_142to173 = f142to174_train[all_imputation_cols].median()

for col in all_imputation_cols:
    # 1) fillna by customer‐group median using transform on a single column
    filled = (
        f142to174_train
        .groupby('id2')[col]
        .transform(lambda s: s.fillna(s.median()))
    )
    # 2) fill any remaining with the global median for that col
    f142to174_train[col] = filled.fillna(global_medians_train_142to173[col])

# Check null percentages after imputation for training data
print("Null % after imputation for f142to174_train:")
print(f142to174_train[all_imputation_cols].isnull().mean() * 100)
```
