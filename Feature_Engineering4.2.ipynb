```python
import os
import pandas as pd

folder_path = '/content/drive/MyDrive/Transformed_Train_Data'

# Get a list of all files and directories in the folder
items_in_folder = os.listdir(folder_path)

# Iterate through each item
loaded_data = {}
for item_name in items_in_folder:
    item_path = os.path.join(folder_path, item_name)

    # Skip if it's a directory
    if os.path.isdir(item_path):
        print(f"Skipping directory: {item_name}")
        continue

    try:
        # Try reading as a parquet file first
        if item_name.endswith('.parquet'):
            df = pd.read_parquet(item_path)
            print(f"Shape of {item_name}: {df.shape}")
            loaded_data[item_name] = df
        # Try reading as a CSV with different encodings if parquet fails or it's not a parquet file
        elif item_name.endswith('.csv'):
             try:
                 df = pd.read_csv(item_path)
                 print(f"Shape of {item_name}: {df.shape}")
                 loaded_data[item_name] = df
             except Exception as csv_error:
                 try:
                     df = pd.read_csv(item_path, encoding='latin1')
                     print(f"Shape of {item_name} (latin1 encoding): {df.shape}")
                     loaded_data[item_name] = df
                 except Exception as latin1_error:
                     print(f"Could not load {item_name} with utf-8 or latin1 encoding: {latin1_error}")
        else:
            print(f"Skipping unknown file type: {item_name}")

    except Exception as e:
        print(f"Could not load {item_name}: {e}")
```

**Reasoning**:
Impute missing values in the specified columns of the training data by grouping by 'id3' and using the median, with a fallback to the global median.



```python
# Create a copy to avoid modifying the original dataframe
f130to141_train_imputed_id3 = f130to141_train_modified.copy()

# Impute missing values in id3_grouped_cols
for col in id3_grouped_cols:
    # Calculate the global median for the current column
    global_median = f130to141_train_imputed_id3[col].median()

    # Group by 'id3' and fill missing values within each group with the group median
    # Use transform to return a Series with the same index as the original DataFrame
    f130to141_train_imputed_id3[col] = f130to141_train_imputed_id3.groupby('id3')[col].transform(
        lambda x: x.fillna(x.median())
    )

    # Fill any remaining missing values with the global median
    f130to141_train_imputed_id3[col] = f130to141_train_imputed_id3[col].fillna(global_median)

# Display the head and check for null values in the imputed columns
print("Head of f130to141_train_imputed_id3 after imputation:")
display(f130to141_train_imputed_id3.head())

print("\nChecking for null values in imputed columns (id3 grouped):")
print(f130to141_train_imputed_id3[id3_grouped_cols].isnull().sum())
```

**Reasoning**:
Impute missing values in columns f140-f141 of dataframe f130to141_train_imputed_id3 by grouping by 'id2' and using the median of non-null values within each group, falling back to the global median if a group has no non-null values.



```python
# Create a copy of the DataFrame
f130to141_train_imputed_id2_id3 = f130to141_train_imputed_id3.copy()

# Define the columns to be imputed in this subtask
id2_grouped_cols = ['f140', 'f141']

# Iterate through each column and perform grouped median imputation with global median fallback
for col in id2_grouped_cols:
    # Calculate the global median for the column
    global_median = f130to141_train_imputed_id2_id3[col].median()

    # Impute missing values by grouping by 'id2' and using the median of non-null values,
    # falling back to the global median if a group has no non-null values.
    f130to141_train_imputed_id2_id3[col] = f130to141_train_imputed_id2_id3.groupby('id2')[col].transform(
        lambda x: x.fillna(x.median() if pd.notna(x.median()) else global_median)
    )

# Display the head of the imputed DataFrame
print("Head of f130to141_train_imputed_id2_id3 after imputation:")
display(f130to141_train_imputed_id2_id3.head())

# Print the count of missing values for each column to verify imputation
print("\nMissing values count per column after imputation:")
print(f130to141_train_imputed_id2_id3[id2_grouped_cols].isnull().sum())
```
