```python
from xgboost import XGBClassifier
from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold
from sklearn.metrics import roc_auc_score, log_loss
from scipy.stats import uniform, randint

xgb_model = XGBClassifier(objective='binary:logistic', use_label_encoder=False, eval_metric='logloss', n_jobs=-1, random_state=42)

# Convert param_grid to a parameter distribution for RandomizedSearchCV
param_distributions = {
    'learning_rate': uniform(loc=0.01, scale=0.19), # Example distribution based on [0.01, 0.1, 0.2]
    'max_depth': randint(low=3, high=8), # Example distribution based on [3, 5, 7]
    'n_estimators': randint(low=100, high=301), # Example distribution based on [100, 200, 300]
    'subsample': uniform(loc=0.8, scale=0.2), # Example distribution based on [0.8, 1.0]
    'colsample_bytree': uniform(loc=0.8, scale=0.2), # Example distribution based on [0.8, 1.0]
}


cv_strategy = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)

scoring_metrics = {
    'AUC': 'roc_auc',
    'LogLoss': 'neg_log_loss'
}

# Use RandomizedSearchCV instead of GridSearchCV
random_search = RandomizedSearchCV(
    xgb_model,
    param_distributions,
    n_iter=50, # Number of parameter settings that are sampled. You can adjust this.
    cv=cv_strategy,
    scoring=scoring_metrics,
    refit='AUC',  # Refit based on AUC, as it was the initial request, but LogLoss is tracked
    return_train_score=True,
    random_state=42,
    n_jobs=-1 # Use all available cores
)

# Note: Early stopping is applied during the random_search.fit() call by passing eval_set and early_stopping_rounds
# For now, the setup is complete, and the next step will be to fit the model.
```
