```python
import pandas as pd

add_trans = pd.read_parquet('/content/drive/MyDrive/AMEXChallenge/add_trans.parquet')
display(add_trans.head())
```

```python
max_f370 = add_trans['f370'].max()
min_f370 = add_trans['f370'].min()

print(f"Maximum value of f370: {max_f370}")
print(f"Minimum value of f370: {min_f370}")
```

```python
train_dataset = pd.read_parquet('/content/drive/MyDrive/AMEXChallenge/train_data.parquet')
display(train_dataset.head())
```

```python
max_id5 = train_dataset['id5'].max()
min_id5 = train_dataset['id5'].min()

print(f"Maximum value of id5: {max_id5}")
print(f"Minimum value of id5: {min_id5}")
```

```python
add_trans_filtered = add_trans[(add_trans['f370'] >= min_id5) & (add_trans['f370'] <= max_id5)]
display(add_trans_filtered.head())
```

```python
print(f"The size of the filtered dataset is: {add_trans_filtered.shape}")
```

```python
null_percentage = add_trans_filtered.isnull().sum() / len(add_trans_filtered) * 100
display(null_percentage)
```

```python
transaction = add_trans_filtered
display(transaction.head())
```

```python
aggregated_transaction = transaction.groupby('id2').agg(
    total_f367=('f367', 'sum'),
    count_D_f369=('f369', lambda x: (x == 'D').sum()),
    count_C_f369=('f369', lambda x: (x == 'C').sum())
).reset_index()

display(aggregated_transaction.head())
```

```python
unique_f374 = transaction['f374'].unique()
print("Unique values in f374:")
print(unique_f374)
```

```python
from sklearn.preprocessing import LabelEncoder

# Calculate the mode for f368 and f374 for each id2
modes = transaction.groupby('id2').agg(
    mode_f368=('f368', lambda x: x.mode()[0] if not x.mode().empty else None),
    mode_f374=('f374', lambda x: x.mode()[0] if not x.mode().empty else None)
).reset_index()

# Merge the modes with the aggregated_transaction dataframe
aggregated_transaction = pd.merge(aggregated_transaction, modes, on='id2', how='left')

# Label encode the mode columns
label_encoder_f368 = LabelEncoder()
aggregated_transaction['mode_f368_encoded'] = label_encoder_f368.fit_transform(aggregated_transaction['mode_f368'])

label_encoder_f374 = LabelEncoder()
aggregated_transaction['mode_f374_encoded'] = label_encoder_f374.fit_transform(aggregated_transaction['mode_f374'])

# Drop the original mode columns
aggregated_transaction = aggregated_transaction.drop(columns=['mode_f368', 'mode_f374'])

display(aggregated_transaction.head())
```

```python
columns_to_keep = [
    'id2',
    'total_f367',
    'count_D_f369',
    'count_C_f369',
    'mode_f368_encoded',
    'mode_f374_encoded'
]

# Select only the specified columns
aggregated_transaction = aggregated_transaction[columns_to_keep]

display(aggregated_transaction.head())
```

```python
aggregated_transaction["total_txn_3d"] = aggregated_transaction["count_D_f369"] + aggregated_transaction["count_C_f369"]
aggregated_transaction["debit_ratio"] = aggregated_transaction["count_D_f369"] / (aggregated_transaction["total_txn_3d"] + 1e-5)  # avoid divide-by-zero
aggregated_transaction["avg_txn_amt_3d"] = aggregated_transaction["total_f367"] / (aggregated_transaction["total_txn_3d"] + 1e-5)  # avoid divide-by-zero

display(aggregated_transaction.head())
```

```python
for col in aggregated_transaction.columns:
    print(f"Column: {col}")
    if pd.api.types.is_numeric_dtype(aggregated_transaction[col]):
        print(f"  Range: ({aggregated_transaction[col].min()}, {aggregated_transaction[col].max()})")
    else:
        print(f"  Unique values: {aggregated_transaction[col].unique()[:10]}...") # Displaying only first 10 unique values for non-numeric
    print(f"  Null Percentage: {aggregated_transaction[col].isnull().sum() / len(aggregated_transaction) * 100:.2f}%")
    print("-" * 30)
```

```python
for col in ['total_f367', 'avg_txn_amt_3d']:
    # Calculate the 1st and 99th percentiles
    lower_bound = aggregated_transaction[col].quantile(0.01)
    upper_bound = aggregated_transaction[col].quantile(0.99)

    # Apply trimming
    aggregated_transaction[col] = aggregated_transaction[col].clip(lower=lower_bound, upper=upper_bound)

display(aggregated_transaction.head())
```
