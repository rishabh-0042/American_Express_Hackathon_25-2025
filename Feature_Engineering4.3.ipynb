```python
import pandas as pd

file_path_1 = '/content/drive/MyDrive/Transformed_Test_Data/f104to309_test.parquet'
file_path_2 = '/content/drive/MyDrive/Transformed_Test_Data/f94to103_test.parquet'

df_f104to309 = pd.read_parquet(file_path_1)
df_f94to103 = pd.read_parquet(file_path_2)

print("DataFrame from f104to309_test.parquet:")
display(df_f104to309.head())

print("\nDataFrame from f94to103_test.parquet:")
display(df_f94to103.head())
```

```python
null_percentages = (df_f94to103.isnull().sum() / len(df_f94to103)) * 100
print("Percentage of null values in df_f94to103:")
display(null_percentages)
```

```python
file_path_train = '/content/drive/MyDrive/Transformed_Train_Data/f94to103_train.parquet'
f94to103_train = pd.read_parquet(file_path_train)

# Rename the existing DataFrame
f94to103_test = df_f94to103

print("DataFrame from f94to103_train.parquet:")
display(f94to103_train.head())

print("\nDataFrame f94to103_test (renamed from df_f94to103):")
display(f94to103_test.head())
```

```python
print("Percentage of null values in f94to103_train:")
display((f94to103_train.isnull().sum() / len(f94to103_train)) * 100)

print("\nPercentage of null values in f94to103_test:")
display((f94to103_test.isnull().sum() / len(f94to103_test)) * 100)
```

```python
# Columns to impute (excluding 'id2')
cols_to_impute_train = f94to103_train.columns[f94to103_train.isnull().any()].tolist()
if 'id2' in cols_to_impute_train:
    cols_to_impute_train.remove('id2')

cols_to_impute_test = f94to103_test.columns[f94to103_test.isnull().any()].tolist()
if 'id2' in cols_to_impute_test:
    cols_to_impute_test.remove('id2')

# Ensure both dataframes have the same columns to impute
cols_to_impute = list(set(cols_to_impute_train + cols_to_impute_test))
if 'id2' in cols_to_impute:
    cols_to_impute.remove('id2')

print(f"Columns to impute: {cols_to_impute}")

# Convert columns to numeric, coercing errors
for col in cols_to_impute:
    f94to103_train[col] = pd.to_numeric(f94to103_train[col], errors='coerce')
    f94to103_test[col] = pd.to_numeric(f94to103_test[col], errors='coerce')

# Calculate global medians for fallback
global_medians_train = f94to103_train[cols_to_impute].median()
global_medians_test = f94to103_test[cols_to_impute].median()


# Impute null values
imputed_cols_train = {}
imputed_cols_test = {}

for col in cols_to_impute:
    print(f"Imputing column: {col}")

    # Imputation for train data
    f94to103_train[col] = f94to103_train.groupby('id2')[col].transform(lambda x: x.fillna(x.median()))
    # Fallback to global median for remaining nulls in train data
    f94to103_train[col].fillna(global_medians_train[col], inplace=True)
    if f94to103_train[col].isnull().sum() == 0:
        imputed_cols_train[col] = True
    else:
        imputed_cols_train[col] = False
        print(f"Warning: Nulls remaining in train data for column {col}")


    # Imputation for test data
    f94to103_test[col] = f94to103_test.groupby('id2')[col].transform(lambda x: x.fillna(x.median()))
    # Fallback to global median for remaining nulls in test data
    f94to103_test[col].fillna(global_medians_test[col], inplace=True)
    if f94to103_test[col].isnull().sum() == 0:
        imputed_cols_test[col] = True
    else:
        imputed_cols_test[col] = False
        print(f"Warning: Nulls remaining in test data for column {col}")

print("\nNull percentages after imputation in f94to103_train:")
display((f94to103_train.isnull().sum() / len(f94to103_train)) * 100)

print("\nNull percentages after imputation in f94to103_test:")
display((f94to103_test.isnull().sum() / len(f94to103_test)) * 100)

print("\nImputation status for train data:")
display(imputed_cols_train)

print("\nImputation status for test data:")
display(imputed_cols_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Columns to scale (excluding 'id2')
cols_to_scale = [col for col in f94to103_train.columns if col != 'id2']

# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Apply log1p and then Min-Max scaling to each column in train and test data
for col in cols_to_scale:
    print(f"Applying log1p and scaling to column: {col}")

    # Apply log1p transformation
    f94to103_train[col] = np.log1p(f94to103_train[col])
    f94to103_test[col] = np.log1p(f94to103_test[col])

    # Apply Min-Max scaling
    f94to103_train[col] = scaler.fit_transform(f94to103_train[[col]])
    f94to103_test[col] = scaler.transform(f94to103_test[[col]])

print("\nTransformed and scaled f94to103_train:")
display(f94to103_train.head())

print("\nTransformed and scaled f94to103_test:")
display(f94to103_test.head())
```

```python
print("Value range (Min and Max) for f94to103_train:")
display(f94to103_train.agg(['min', 'max']))

print("\nValue range (Min and Max) for f94to103_test:")
display(f94to103_test.agg(['min', 'max']))
```

```python
# Identify columns with null values (excluding 'id2')
cols_to_impute_train = f1to41_train.columns[f1to41_train.isnull().any()].tolist()
if 'id2' in cols_to_impute_train:
    cols_to_impute_train.remove('id2')

cols_to_impute_test = f1to41_test.columns[f1to41_test.isnull().any()].tolist()
if 'id2' in cols_to_impute_test:
    cols_to_impute_test.remove('id2')

# Combine columns to impute from both dataframes
cols_to_impute = list(set(cols_to_impute_train + cols_to_impute_test))
if 'id2' in cols_to_impute:
    cols_to_impute.remove('id2')

print(f"Columns to impute: {cols_to_impute}")

# Calculate global medians for fallback
global_medians_train = f1to41_train[cols_to_impute].median()
global_medians_test = f1to41_test[cols_to_impute].median()

# Impute null values and track status
imputed_cols_train = {}
imputed_cols_test = {}

for col in cols_to_impute:
    print(f"Imputing column: {col}")

    # Imputation for train data
    f1to41_train[col] = f1to41_train.groupby('id2')[col].transform(lambda x: x.fillna(x.median()))
    # Fallback to global median for remaining nulls in train data
    f1to41_train[col].fillna(global_medians_train[col], inplace=True)
    if f1to41_train[col].isnull().sum() == 0:
        imputed_cols_train[col] = True
    else:
        imputed_cols_train[col] = False
        print(f"Warning: Nulls remaining in train data for column {col}")

    # Imputation for test data
    f1to41_test[col] = f1to41_test.groupby('id2')[col].transform(lambda x: x.fillna(x.median()))
    # Fallback to global median for remaining nulls in test data
    f1to41_test[col].fillna(global_medians_test[col], inplace=True)
    if f1to41_test[col].isnull().sum() == 0:
        imputed_cols_test[col] = True
    else:
        imputed_cols_test[col] = False
        print(f"Warning: Nulls remaining in test data for column {col}")

print("\nNull percentages after imputation in f1to41_train:")
display((f1to41_train.isnull().sum() / len(f1to41_train)) * 100)

print("\nNull percentages after imputation in f1to41_test:")
display((f1to41_test.isnull().sum() / len(f1to41_test)) * 100)

print("\nImputation status for train data:")
display(imputed_cols_train)

print("\nImputation status for test data:")
display(imputed_cols_test)
```

```python
from sklearn.preprocessing import MinMaxScaler, RobustScaler
import numpy as np

# Define column groups - Ensure these match the columns present in f1to41_train and f1to41_test
# Check the columns in the current dataframes to confirm the groups
current_cols_train = f1to41_train.columns.tolist()
current_cols_test = f1to41_test.columns.tolist()

# Assuming the columns dropped previously are still dropped: f14, f15, f16, f17, f19, f20, f21, f23, f24, f25
# Adjust column groups based on the remaining columns in the current dataframes
cols_group1 = [col for col in ['f1', 'f2', 'f3', 'f4', 'f5', 'f6', 'f7', 'f8', 'f9', 'f10', 'f11', 'f12', 'f13', 'f18', 'f32', 'f34', 'f37'] if col in current_cols_train]
cols_group2 = [col for col in ['f22', 'f26', 'f27', 'f28', 'f29', 'f30', 'f31', 'f33', 'f35', 'f36', 'f38'] if col in current_cols_train]
cols_group3 = [col for col in ['f39', 'f40', 'f41'] if col in current_cols_train]


print(f"Columns for MinMaxScaler (Group 1): {cols_group1}")
print(f"Columns for log1p + MinMaxScaler (Group 2): {cols_group2}")
print(f"Columns for Winsorize + RobustScaler (Group 3): {cols_group3}")


# Apply MinMaxScaler (Group 1)
print("\nApplying MinMaxScaler to Group 1...")
if cols_group1: # Check if group is not empty
    scaler_group1 = MinMaxScaler()
    f1to41_train[cols_group1] = scaler_group1.fit_transform(f1to41_train[cols_group1])
    f1to41_test[cols_group1] = scaler_group1.transform(f1to41_test[cols_group1])
    print("MinMaxScaler applied to Group 1.")
    print("NaN check after Group 1 scaling (train):", f1to41_train[cols_group1].isnull().sum().sum())
    print("NaN check after Group 1 scaling (test):", f1to41_test[cols_group1].isnull().sum().sum())
    print("Value range for Group 1 (train):")
    display(f1to41_train[cols_group1].agg(['min', 'max']))
    print("Value range for Group 1 (test):")
    display(f1to41_test[cols_group1].agg(['min', 'max']))
else:
    print("Group 1 is empty, skipping transformation.")


# Apply log1p and MinMaxScaler (Group 2)
print("\nApplying log1p and MinMaxScaler to Group 2...")
if cols_group2: # Check if group is not empty
    for col in cols_group2:
        # Apply log1p transformation
        f1to41_train[col] = np.log1p(f1to41_train[col])
        f1to41_test[col] = np.log1p(f1to41_test[col])
    print("log1p transformation applied to Group 2.")
    print("NaN check after Group 2 log1p (train):", f1to41_train[cols_group2].isnull().sum().sum())
    print("NaN check after Group 2 log1p (test):", f1to41_test[cols_group2].isnull().sum().sum())
    print("Value range for Group 2 after log1p (train):")
    display(f1to41_train[cols_group2].agg(['min', 'max']))
    print("Value range for Group 2 after log1p (test):")
    display(f1to41_test[cols_group2].agg(['min', 'max']))

    scaler_group2 = MinMaxScaler()
    f1to41_train[cols_group2] = scaler_group2.fit_transform(f1to41_train[cols_group2])
    f1to41_test[cols_group2] = scaler_group2.transform(f1to41_test[cols_group2])
    print("MinMaxScaler applied to Group 2.")
    print("NaN check after Group 2 scaling (train):", f1to41_train[cols_group2].isnull().sum().sum())
    print("NaN check after Group 2 scaling (test):", f1to41_test[cols_group2].isnull().sum().sum())
    print("Value range for Group 2 (train):")
    display(f1to41_train[cols_group2].agg(['min', 'max']))
    print("Value range for Group 2 (test):")
    display(f1to41_test[cols_group2].agg(['min', 'max']))
else:
    print("Group 2 is empty, skipping transformation.")


# Apply Winsorize and RobustScaler (Group 3)
print("\nApplying Winsorize and RobustScaler to Group 3...")
if cols_group3: # Check if group is not empty
    for col in cols_group3:
        # Calculate winsorization limits from train data
        lo, hi = f1to41_train[col].quantile([0.01, 0.99])
        # Apply clipping to both train and test data
        f1to41_train[col] = f1to41_train[col].clip(lower=lo, upper=hi)
        f1to41_test[col]  = f1to41_test[col].clip(lower=lo, upper=hi)
    print("Winsorization (clipping) applied to Group 3.")
    print("NaN check after Group 3 clipping (train):", f1to41_train[cols_group3].isnull().sum().sum())
    print("NaN check after Group 3 clipping (test):", f1to41_test[cols_group3].isnull().sum().sum())
    print("Value range for Group 3 after clipping (train):")
    display(f1to41_train[cols_group3].agg(['min', 'max']))
    print("Value range for Group 3 after clipping (test):")
    display(f1to41_test[cols_group3].agg(['min', 'max']))

    scaler_group3 = RobustScaler()
    f1to41_train[cols_group3] = scaler_group3.fit_transform(f1to41_train[cols_group3])
    f1to41_test[cols_group3] = scaler_group3.transform(f1to41_test[cols_group3])
    print("RobustScaler applied to Group 3.")
    print("NaN check after Group 3 scaling (train):", f1to41_train[cols_group3].isnull().sum().sum())
    print("NaN check after Group 3 scaling (test):", f1to41_test[cols_group3].isnull().sum().sum())
    print("Value range for Group 3 (train):")
    display(f1to41_train[cols_group3].agg(['min', 'max']))
    print("Value range for Group 3 (test):")
    display(f1to41_test[cols_group3].agg(['min', 'max']))
else:
    print("Group 3 is empty, skipping transformation.")


print("\nPreprocessing of f1to41 dataframes complete.")
print("\nFirst 5 rows of fully transformed f1to41_train:")
display(f1to41_train.head())

print("\nFirst 5 rows of fully transformed f1to41_test:")
display(f1to41_test.head())
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np

# Columns for additional preprocessing
cols_to_process = ['f39', 'f40', 'f41']

# Apply Winsorization (clipping at 5th and 95th percentiles)
print("Applying Winsorization (clipping at 5th and 95th percentiles)...")
for col in cols_to_process:
    # Calculate clipping limits from train data
    lo, hi = f1to41_train[col].quantile([0.05, 0.95])
    # Apply clipping to both train and test data
    f1to41_train[col] = f1to41_train[col].clip(lower=lo, upper=hi)
    f1to41_test[col] = f1to41_test[col].clip(lower=lo, upper=hi)
print("Winsorization (clipping) applied.")

# Apply StandardScaler
print("\nApplying StandardScaler...")
scaler_spend = StandardScaler()
f1to41_train[cols_to_process] = scaler_spend.fit_transform(f1to41_train[cols_to_process])
f1to41_test[cols_to_process] = scaler_spend.transform(f1to41_test[cols_to_process])
print("StandardScaler applied.")


# Sanity Check: Re-print the min/max of f39â€“f41
print("\nSanity Check: Value range (Min and Max) for 'f39', 'f40', 'f41' after scaling (train):")
display(f1to41_train[cols_to_process].agg(['min', 'max']))
print("\nSanity Check: Value range (Min and Max) for 'f39', 'f40', 'f41' after scaling (test):")
display(f1to41_test[cols_to_process].agg(['min', 'max']))

print("\nAdditional preprocessing for 'f39', 'f40', 'f41' complete.")
```
