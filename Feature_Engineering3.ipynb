```python

import pandas as pd

test_data = pd.read_parquet('/content/drive/MyDrive/AMEXChallenge/test_data.parquet')
train_data = pd.read_parquet('/content/drive/MyDrive/AMEXChallenge/train_data.parquet')

print("Test data loaded successfully.")
print("Train data loaded successfully.")

# Display the first few rows of each DataFrame to verify
print("\nTest data head:")
display(test_data.head())

print("\nTrain data head:")
display(train_data.head())
```

```python
f325to328_train = train_data[['id2', 'f325', 'f326', 'f327', 'f328']]
f325to328_test = test_data[['id2', 'f325', 'f326', 'f327', 'f328']]

print("Train subset created successfully.")
display(f325to328_train.head())

print("\nTest subset created successfully.")
display(f325to328_test.head())
```

```python
import numpy as np

def analyze_dataframe(df, name):
    print(f"Analyzing {name} DataFrame:")
    print("-" * 30)
    for col in df.columns:
        if df[col].dtype in [np.float64, np.int64]:
            null_percentage = df[col].isnull().sum() / len(df) * 100
            print(f"Column '{col}':")
            print(f"  Null Percentage: {null_percentage:.2f}%")
            data_range = f"{df[col].min():.2f} - {df[col].max():.2f}"
            print(f"  Range: {data_range}")
            print("-" * 20)
        else:
            print(f"Column '{col}':")
            print("  Not a numeric data type.")
            print("-" * 20)


analyze_dataframe(f325to328_train, "f325to328_train")
analyze_dataframe(f325to328_test, "f325to328_test")
```

```python
print("Data types for f325to328_train:")
print(f325to328_train.dtypes)

print("\nData types for f325to328_test:")
print(f325to328_test.dtypes)
```

```python
for col in ['f325', 'f326', 'f327', 'f328']:
    f325to328_train[col] = pd.to_numeric(f325to328_train[col], errors='coerce')
    f325to328_test[col] = pd.to_numeric(f325to328_test[col], errors='coerce')

print("Columns converted to float in f325to328_train:")
print(f325to328_train.dtypes)

print("\nColumns converted to float in f325to328_test:")
print(f325to328_test.dtypes)

print("\nFirst few rows of f325to328_train after conversion:")
display(f325to328_train.head())

print("\nFirst few rows of f325to328_test after conversion:")
display(f325to328_test.head())
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Apply log1p transformation
for col in ['f325', 'f326', 'f327', 'f328']:
    # Applying log1p to handle potential zeros and skewed data
    # Add a small constant or handle zeros if necessary before log1p
    # For now, assuming non-negative values for log1p
    f325to328_train[col] = np.log1p(f325to328_train[col])
    f325to328_test[col] = np.log1p(f325to328_test[col])


# Normalize the data using MinMaxScaler
scaler = MinMaxScaler()

# Fit the scaler on the training data and transform both train and test data
# Only apply to the numeric columns
numeric_cols = ['f325', 'f326', 'f327', 'f328']

f325to328_train[numeric_cols] = scaler.fit_transform(f325to328_train[numeric_cols])
f325to328_test[numeric_cols] = scaler.transform(f325to328_test[numeric_cols])


print("Log1p transformation and normalization applied.")

print("\nFirst few rows of f325to328_train after transformation and normalization:")
display(f325to328_train.head())

print("\nFirst few rows of f325to328_test after transformation and normalization:")
display(f325to328_test.head())
```

```python
# Define the columns to impute
cols_to_impute = ['f325', 'f326', 'f327', 'f328']

# Calculate the overall median for each column as a fallback
overall_medians_train = f325to328_train[cols_to_impute].median()
overall_medians_test = f325to328_test[cols_to_impute].median()

# Define a function to impute based on group median with overall median fallback
def impute_with_group_median(group, overall_medians):
    return group.fillna(group.median()).fillna(overall_medians)

# Apply the imputation to the training data
f325to328_train[cols_to_impute] = f325to328_train.groupby('id2')[cols_to_impute].transform(
    lambda x: impute_with_group_median(x, overall_medians_train)
)

# Apply the imputation to the test data
f325to328_test[cols_to_impute] = f325to328_test.groupby('id2')[cols_to_impute].transform(
    lambda x: impute_with_group_median(x, overall_medians_test)
)

print("Missing values imputed based on group median with overall median fallback.")

print("\nFirst few rows of f325to328_train after imputation:")
display(f325to328_train.head())

print("\nFirst few rows of f325to328_test after imputation:")
display(f325to328_test.head())

# Verify if there are any remaining null values in the imputed columns
print("\nNull value check after imputation for f325to328_train:")
print(f325to328_train[cols_to_impute].isnull().sum())

print("\nNull value check after imputation for f325to328_test:")
print(f325to328_test[cols_to_impute].isnull().sum())
```

```python
from sklearn.preprocessing import RobustScaler
import numpy as np

cols_to_process = ['f355', 'f356', 'f357']

# Apply clipping based on percentiles
for col in cols_to_process:
    # Calculate 1st and 99th percentiles from the training data, ignoring NaNs
    lower_bound = f351to357_train[col].quantile(0.01)
    upper_bound = f351to357_train[col].quantile(0.99)

    # Apply clipping to both train and test data, keeping NaNs
    f351to357_train[col] = f351to357_train[col].clip(lower=lower_bound, upper=upper_bound)
    f351to357_test[col] = f351to357_test[col].clip(lower=lower_bound, upper=upper_bound)

print("Clipping applied to f355, f356, and f357 based on 1st and 99th percentiles.")

# Initialize the RobustScaler
scaler = RobustScaler()

# Apply RobustScaler
# Fit the scaler on the training data and transform both train and test data
# RobustScaler handles NaNs by default, but clipping was done in place while preserving NaNs
f351to357_train[cols_to_process] = scaler.fit_transform(f351to357_train[cols_to_process])
f351to357_test[cols_to_process] = scaler.transform(f351to357_test[cols_to_process])

print("RobustScaler applied to f355, f356, and f357.")

# Analyze the transformed and scaled columns
print("\nAnalysis after clipping and RobustScaler:")
analyze_dataframe(f351to357_train, "f351to357_train after clipping and RobustScaler")
analyze_dataframe(f351to357_test, "f351to357_test after clipping and RobustScaler")

print("\nFirst few rows of f351to357_train after clipping and RobustScaler:")
display(f351to357_train.head())

print("\nFirst few rows of f351to357_test after clipping and RobustScaler:")
display(f351to357_test.head())
```

```python
cols_to_impute = ['f351', 'f352', 'f353', 'f355', 'f356', 'f357']

# Calculate the overall median for each column as a fallback
overall_medians_train = f351to357_train[cols_to_impute].median()
overall_medians_test = f351to357_test[cols_to_impute].median()

# Apply group-wise median imputation
f351to357_train[cols_to_impute] = f351to357_train.groupby('id3')[cols_to_impute].transform(lambda x: x.fillna(x.median()))
f351to357_test[cols_to_impute] = f351to357_test.groupby('id3')[cols_to_impute].transform(lambda x: x.fillna(x.median()))

# Apply overall median imputation for any remaining NaNs
f351to357_train[cols_to_impute] = f351to357_train[cols_to_impute].fillna(overall_medians_train)
f351to357_test[cols_to_impute] = f351to357_test[cols_to_impute].fillna(overall_medians_test)


print("Missing values in f351 to f353 and f355 to f357 imputed using group median with overall median fallback.")

# Verify if there are any remaining null values in the imputed columns
print("\nNull value check after imputation for f351to357_train:")
print(f351to357_train[cols_to_impute].isnull().sum())

print("\nNull value check after imputation for f351to357_test:")
print(f351to357_test[cols_to_impute].isnull().sum())

print("\nFirst few rows of f351to357_train after imputation:")
display(f351to357_train.head())

print("\nFirst few rows of f351to357_test after imputation:")
display(f351to357_test.head())
```
