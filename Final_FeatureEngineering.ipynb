```python
import pandas as pd

transaction = pd.read_parquet('/content/drive/MyDrive/Colab Notebooks/AMEXChallenge/add_trans.parquet')
offer = pd.read_parquet('/content/drive/MyDrive/Colab Notebooks/AMEXChallenge/offer_metadata.parquet')

print("Transaction DataFrame head:")
display(transaction.head())

print("\nOffer DataFrame head:")
display(offer.head())
```

```python
train_df = pd.read_parquet('/content/drive/MyDrive/Colab Notebooks/AMEXChallenge/train_data.parquet')
print("\nTrain DataFrame head:")
display(train_df.head())
```

```python
min_id5 = train_df['id5'].min()
max_id5 = train_df['id5'].max()

print(f"The minimum value of 'id5' is: {min_id5}")
print(f"The maximum value of 'id5' is: {max_id5}")
```

```python
transaction['f370'] = pd.to_datetime(transaction['f370'])

filtered_transaction = transaction[(transaction['f370'] >= min_id5) & (transaction['f370'] <= max_id5)]

print("Filtered Transaction DataFrame head:")
display(filtered_transaction.head())
```

```python
# Check data types
print(f"Data type of 'id8' in filtered_transaction: {filtered_transaction['id8'].dtype}")
print(f"Data type of 'id8' in offer: {offer['id8'].dtype}")

# Convert to int64, coercing errors to NaN
try:
    filtered_transaction['id8'] = pd.to_numeric(filtered_transaction['id8'], errors='coerce').astype('Int64')
    offer['id8'] = pd.to_numeric(offer['id8'], errors='coerce').astype('Int64')
    print("\nConverted 'id8' columns to Int64.")
except Exception as e:
    print(f"\nError during conversion: {e}")


# Find matching id8 values, dropping NaN values created during conversion
matching_id8_count = filtered_transaction['id8'].dropna().isin(offer['id8'].dropna()).sum()

print(f"\nNumber of matching 'id8' values between filtered_transaction and offer: {matching_id8_count}")
```

```python
print("Shape of offer DataFrame:", offer.shape)
print("Shape of filtered_transaction DataFrame:", filtered_transaction.shape)
```

```python
# Group by id8 and calculate the sum of f367 and frequencies of 'D' and 'C' in f369
grouped_transaction = filtered_transaction.groupby('id8').agg(
    total_f367=('f367', 'sum'),
    freq_D_f369=('f369', lambda x: (x == 'D').sum()),
    freq_C_f369=('f369', lambda x: (x == 'C').sum())
).reset_index()

print("Grouped Transaction DataFrame head:")
display(grouped_transaction.head())
```

```python
print("Shape of grouped_transaction DataFrame:", grouped_transaction.shape)
```

```python
unique_f375 = offer['f375'].unique()
print(f"Unique values in 'f375': {unique_f375}")

min_f376 = offer['f376'].min()
max_f376 = offer['f376'].max()
print(f"Range of values in 'f376': {min_f376} to {max_f376}")
```

```python
offer['id12'] = pd.to_datetime(offer['id12'])
offer['id13'] = pd.to_datetime(offer['id13'])

filtered_offer = offer[(offer['id12'] <= '2023-11-03') & (offer['id13'] >= '2023-11-01')]

print("Filtered Offer DataFrame head:")
display(filtered_offer.head())
```

```python
# Function to find the mode, handling potential empty results after dropping NaNs
def mode_or_none(x):
    mode_values = x.dropna().mode()
    if not mode_values.empty:
        return mode_values.iloc[0]
    else:
        return None

# Group by id3 and calculate the required aggregations
grouped_offer = filtered_offer.groupby('id3').agg(
    median_f376=('f376', lambda x: x.dropna().median()),
    freq_f375_2=('f375', lambda x: (x == 2).sum()),
    freq_f375_1=('f375', lambda x: (x == 1).sum()),
    most_frequent_id8=('id8', mode_or_none)
).reset_index()

print("Grouped Offer DataFrame head:")
display(grouped_offer.head())
```

```python
print("Shape of grouped_offer DataFrame:", grouped_offer.shape)
print("Data type of 'id3' in grouped_offer:", grouped_offer['id3'].dtype)
```

```python
num_unique_id3 = filtered_offer['id3'].nunique()
print(f"Number of unique 'id3' values in filtered_offer: {num_unique_id3}")
```

```python
null_percentages = filtered_offer.isnull().sum() / len(filtered_offer) * 100
print("Percentage of null values in filtered_offer:")
print(null_percentages)
```

```python
null_percentages_grouped_offer = grouped_offer.isnull().sum() / len(grouped_offer) * 100
print("Percentage of null values in grouped_offer:")
print(null_percentages_grouped_offer)
```

```python
# Calculate the global median for each column
global_medians = grouped_offer.median()

# Fill NaN values with the global median
grouped_offer_filled = grouped_offer.fillna(global_medians)

print("Grouped Offer DataFrame after filling NaNs with global median:")
display(grouped_offer_filled.head())

print("\nPercentage of null values after filling:")
print(grouped_offer_filled.isnull().sum() / len(grouped_offer_filled) * 100)
```

```python
# Merge grouped_offer_filled and grouped_transaction on id8
merged_df = pd.merge(grouped_offer_filled, grouped_transaction, left_on='most_frequent_id8', right_on='id8', how='left')

print("Merged DataFrame head:")
display(merged_df.head())

print("\nShape of the merged DataFrame:", merged_df.shape)
```

```python
null_percentages_merged_df = merged_df.isnull().sum() / len(merged_df) * 100
print("Percentage of null values in merged_df:")
print(null_percentages_merged_df)
```

```python
num_unique_most_frequent_id8 = grouped_offer['most_frequent_id8'].nunique(dropna=True)
print(f"Number of unique non-null 'most_frequent_id8' values in grouped_offer: {num_unique_most_frequent_id8}")
```

```python
# Get the non-null 'most_frequent_id8' values from grouped_offer_filled
most_frequent_id8_in_offer = grouped_offer_filled['most_frequent_id8'].dropna()

# Get the non-null 'id8' values from grouped_transaction
id8_in_transaction = grouped_transaction['id8'].dropna()

# Find the intersection of these two sets of IDs
common_id8_count = most_frequent_id8_in_offer.isin(id8_in_transaction).sum()

print(f"Number of common 'id8' values between grouped_offer_filled and grouped_transaction: {common_id8_count}")
```

```python
merged_df = merged_df.drop('id8', axis=1)

print("Merged DataFrame after dropping 'id8' column:")
display(merged_df.head())

print("\nColumns of the merged DataFrame:")
print(merged_df.columns)
```

```python
# Create a flag column for rows with nulls in the specified columns
cols_to_check = ['total_f367', 'freq_D_f369', 'freq_C_f369']
merged_df['flag_nulls'] = merged_df[cols_to_check].isnull().any(axis=1).astype(int)

# Impute null values in the specified columns with 0
merged_df[cols_to_check] = merged_df[cols_to_check].fillna(0)

print("Merged DataFrame after creating flag and imputing nulls:")
display(merged_df.head())

print("\nPercentage of null values after imputation:")
print(merged_df.isnull().sum() / len(merged_df) * 100)

print("\nValue counts for the new 'flag_nulls' column:")
print(merged_df['flag_nulls'].value_counts())
```

```python
# Convert id4 to datetime
final_train['id4'] = pd.to_datetime(final_train['id4'])
final_test['id4'] = pd.to_datetime(final_test['id4'])

# Extract temporal features for final_train
final_train['hour'] = final_train['id4'].dt.hour
final_train['minute'] = final_train['id4'].dt.minute
final_train['day'] = final_train['id4'].dt.day
final_train['weekday'] = final_train['id4'].dt.weekday
final_train['is_weekend'] = (final_train['id4'].dt.weekday >= 5).astype(int)

# Extract temporal features for final_test
final_test['hour'] = final_test['id4'].dt.hour
final_test['minute'] = final_test['id4'].dt.minute
final_test['day'] = final_test['id4'].dt.day
final_test['weekday'] = final_test['id4'].dt.weekday
final_test['is_weekend'] = (final_test['id4'].dt.weekday >= 5).astype(int)

# Define the get_part_of_day function (already provided by the user)
def get_part_of_day(hour):
    if 5 <= hour < 12:
        return "morning"
    elif 12 <= hour < 17:
        return "afternoon"
    elif 17 <= hour < 21:
        return "evening"
    else:
        return "night"

# Create the part_of_day feature for final_train and final_test
final_train['part_of_day'] = final_train['hour'].apply(get_part_of_day)
final_test['part_of_day'] = final_test['hour'].apply(get_part_of_day)

# Convert part_of_day to categorical dtype
final_train['part_of_day'] = final_train['part_of_day'].astype('category')
final_test['part_of_day'] = final_test['part_of_day'].astype('category')


print("Null percentages of new temporal features in final_train:")
print(final_train[['hour', 'minute', 'day', 'weekday', 'is_weekend', 'part_of_day']].isnull().sum() / len(final_train) * 100)

print("\nNull percentages of new temporal features in final_test:")
print(final_test[['hour', 'minute', 'day', 'weekday', 'is_weekend', 'part_of_day']].isnull().sum() / len(final_test) * 100)

print("\nFinal shape of final_train DataFrame:", final_train.shape)
print("Final shape of final_test DataFrame:", final_test.shape)
```
